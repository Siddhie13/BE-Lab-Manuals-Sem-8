```python
from xmlrpc.server import SimpleXMLRPCServer
import threading
import math

# 1. Define the remote procedure (factorial calculation)
def calculate_factorial(n):
    if not isinstance(n, int):
        return "Error: Input must be an integer."
    if n < 0:
        return "Error: Factorial is not defined for negative numbers."
    
    print(f"[Server] Received request to calculate factorial of {n}")
    result = math.factorial(n)
    print(f"[Server] Calculated result: {result}")
    return result

# 2. Set up the XML-RPC Server
host = "localhost"
port = 8000
server = SimpleXMLRPCServer((host, port), allow_none=True, logRequests=False)

# 3. Register the function so the client can call it
server.register_function(calculate_factorial, "factorial")

# 4. Create a function to run the server
def run_server():
    print(f"RPC Server is running on http://{host}:{port}...")
    server.serve_forever()

# 5. Start the server in a daemon thread so it doesn't block the notebook
server_thread = threading.Thread(target=run_server)
server_thread.daemon = True # Ensures the thread closes when the notebook kernel stops
server_thread.start()
```

```python
import xmlrpc.client

# 1. Connect to the RPC server
server_url = "http://localhost:8000/"
proxy = xmlrpc.client.ServerProxy(server_url)

# 2. Define the integer you want to send
number_to_submit = 5  # You can change this number to test different inputs

print(f"[Client] Submitting number {number_to_submit} to the server...")

# 3. Call the remote procedure and print the result
try:
    # This calls the 'calculate_factorial' function residing on the server
    result = proxy.factorial(number_to_submit)
    print(f"[Client] Result received from server: {number_to_submit}! = {result}")
except Exception as e:
    print(f"[Client] An error occurred: {e}")
```

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt
```

```python
# Load dataset
df = pd.read_csv('Decision_Tree_Dataset.csv')
```

```python
df.head()
```

```python
# Features and target
X = df.drop('Result', axis=1)
y = df['Result']
```

```python
# Create model
model = DecisionTreeClassifier(
    criterion='entropy',     # or 'gini'
    max_depth=3,
    min_samples_split=2,
    random_state=42
)
```

```python
# Train model
model.fit(X, y)
```

```python
# Plot tree
plt.figure(figsize=(20,12), dpi=200)

plot_tree(
    model,
    feature_names=X.columns,
    class_names=['No', 'Yes'],
    filled=True,
    fontsize=12
)

plt.show()
```
